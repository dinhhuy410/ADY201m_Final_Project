```python
import sys, os
sys.path.insert(0, os.path.dirname(os.getcwd()))

import pandas as pd
from statsmodels.tsa.stattools import adfuller
from load_data import load_vcb_data

df = load_vcb_data()

def check_stationarity(series, name):
    print(f"--- Kiểm định ADF cho chuỗi: {name} ---")
    result = adfuller(series.dropna())
    print(f'ADF Statistic: {result[0]:.4f}')
    print(f'p-value: {result[1]:.4f}')
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'   {key}: {value:.4f}')
    if result[1] <= 0.05:
        print("=> Kết luận: Chuỗi ĐÃ DỪNG (Bác bỏ H0 tại mức ý nghĩa 5%)")
    else:
        print("=> Kết luận: Chuỗi KHÔNG DỪNG (Chưa thể bác bỏ H0)")
    print("-" * 40)

# Kiểm tra mức giá đóng cửa tuyệt đối và chuỗi lợi nhuận (pct_change)
check_stationarity(df['Close'], 'Giá Đóng Cửa (Close)')
check_stationarity(df['Close'].pct_change(), 'Tỷ suất sinh lời Ngày (Daily Return)')
```
